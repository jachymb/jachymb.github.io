```python
from scipy.stats import nbinom, poisson, gamma
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 18

d = [7,3,4,4,6,2,4,0,4,1]
n=len(d)
s=sum(d)
# Negative Binomial Parameters
#alpha_vals = [1,2,3,3]  # alpha values
#beta_vals = [1,1,1,0.5]  # beta values
x = np.arange(0, 13)  # Values of x (integer k)

# Plotting the PMFs for different alpha, beta combinations
plt.figure(figsize=(10, 6))
width = 0.1  # Width of the bars
aa = 0.2
offsets = np.arange(len(alpha_vals)) * width-aa  # Offset for each group of bars

for i, (alpha, beta, c) in enumerate(zip(alpha_vals, beta_vals, "rgbm")):
    r = alpha + s
    p = (beta+n)/(beta+n+1)
    #r = alpha
    #p = 1 - beta  # success probability p = 1 - beta
    pmf = nbinom.pmf(x, r, p)
    plt.bar(x+ offsets[i], pmf, width = width, label=f'alpha={alpha}, beta={beta}', color=c)

plt.bar(x+ 0.4-aa, poisson.pmf(x, 3.5), label=f'Poisson(3.5)', width = width, color='k')


# Customizing the plot
plt.title('Negative binomial predictive posterior')
plt.xlabel('x')
plt.ylabel('p(x|X)')
plt.xticks(np.arange(0, max(x)+1, 1))  # Integer ticks on x-axis
plt.legend()
plt.show()
```

```python
# Parameters for the Gamma distribution
alpha_vals = [1, 4*14, 2, 42]  # shape parameters (alpha)
beta_vals = [10, 4*4, 1/5, 6]    # scale parameters (beta)

# X values for plotting
x = np.linspace(0, 12, 1000)  # Range of x values

# Plotting the PDFs for different alpha, beta combinations
plt.figure(figsize=(10, 6))

for alpha, beta, c in zip(alpha_vals, beta_vals, "rgbm"):
    pdf = gamma.pdf(x, a=alpha, scale=1/beta)
    plt.plot(x, pdf, label=f'α={alpha}, β={beta}', color=c)

# Customizing the plot
plt.title('Gamma prior')
plt.xlabel('λ')
plt.ylabel('p(λ)')
plt.ylim(0, 1.4)
#plt.grid(True)
plt.xticks(range(11))
plt.legend()
plt.show()
```

```python
# Parameters for the Gamma distribution

# X values for plotting
x = np.linspace(0, 10, 1000)  # Range of x values

# Plotting the PDFs for different alpha, beta combinations
plt.figure(figsize=(10, 6))

for alpha, beta, c in zip(alpha_vals, beta_vals, "rgbm"):
    pdf = gamma.pdf(x, a=alpha+s, scale=1/(beta+n))
    plt.plot(x, pdf, color=c)

# Customizing the plot
plt.title('Posterior')
plt.xlabel('λ')
plt.ylabel('p(λ|X)')
#plt.xlim(0, 20)
plt.ylim(0, 1.4)
plt.xticks(range(11))
plt.legend()
plt.show()
```

```python
# Parameters for the Gamma distribution

# X values for plotting
x = np.linspace(0, 10, 1000)  # Range of x values

# Plotting the PDFs for different alpha, beta combinations
plt.figure(figsize=(10, 6))

f = np.array([np.product([poisson.pmf(y, lam) for y in d]) for lam in x])
plt.plot(x,f, color='k')

# Customizing the plot
plt.title('Likelihood (not normalized)')
plt.xlabel('λ')
plt.ylabel('p(X|λ)')
#plt.xlim(0, 20)
plt.ylim(0, 5e-10)
plt.xticks(range(11))
plt.legend()
plt.show()
print(sum(f/1000))
```
