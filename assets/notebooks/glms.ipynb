```
SmallestLikelyEvents[p_, lambda_] := Module[  (* this is fucken ugly implementation but meh, just for the plots it work. do not use outside visualization *)
  {probabilities, sortedProbabilities, event, prob, cumulativeSum = 0, selectedEvents = {}},
  
  (* Compute probabilities for a range of k *)
  probabilities = Table[{k, PDF[PoissonDistribution[lambda], k]}, {k, 0, 100}];
  
  (* Sort probabilities by descending likelihood *)
  sortedProbabilities = SortBy[probabilities, -Last[#] &];
  
  (* Find the smallest set such that the total probability > p *)
  Do[
    {event, prob} = ep;
    cumulativeSum += prob;
    AppendTo[selectedEvents, event];
    If[cumulativeSum > p, Break[]],
    {ep, sortedProbabilities}
  ];
  {Min[selectedEvents], Max[selectedEvents]}
];
```
